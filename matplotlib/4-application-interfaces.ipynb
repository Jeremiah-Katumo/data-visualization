#### **-** Matplotlib has two major application interfaces, or styles of using the library:

- An ***explicit*** "Axes" interface that uses methods on a Figure or Axes object to create other Artists, and build a visualization step by step. This has also been called an "object-oriented" interface.

- An ***implicit*** "pyplot" interface that keeps track of the last Figure and Axes created, and adds Artists to the object it thinks the user wants.

```python
import numpy as np
import pandas as plt
import matplotlib.pyplot as plt
```

```python
fig = plt.figure()
ax = fig.subplots()  # ax is created using an instance of the figure using subplots method
ax.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

```python
plt.plot([1, 2, 3, 4], [0, 0.5, 1, 0.2])
```

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [0, 0.5, 0.2])

plt.subplot(1, 2, 2)
plt.plot([3, 2, 1], [0, 0.5, 0.2])
```

```python
plt.subplot(1, 2, 1)
ax = plt.gca()
ax.plot([1,2,3], [0, .5, .2])

plt.subplot(1, 2, 2)
ax = plt.gca()
ax.plot([3,2,1], [0,.5,.2])
```

```python
fig, axs = plt.subplots(1, 2, layout='constrained')
axs[0].plot([1,2,3], [0,.5,.2])
axs[1].plot([3,2,1], [0,.5,.2])
```

```python
plt.subplot(1, 2, 1)
plt.plot([1, 2, 3], [0, 0.5, 0.2])

plt.subplot(1, 2, 2)
plt.plot([3, 2, 1], [0, 0.5, 0.2])

plt.suptitle('Implicit Interface: re-call subplot')

for i in range(1, 3):
    plt.subplot(1, 2, i)
    if i == 1:
        plt.xlabel('Woo')
    else:
        plt.xlabel('Boo')
```

```python
axs = []
ax = plt.subplot(1, 2, 1)
axs += [ax]
plt.plot([1,2,3], [0,.5,.2])

ax = plt.subplot(1, 2, 2)
axs += [ax]
plt.plot([3,2,1], [0,.5,.2])

plt.suptitle("Implicit Interface: save handles")

for i in range(2):
    plt.sca(axs[i])
    if i == 0:
        plt.xlabel('Woo')
    else:
        plt.xlabel('Boo')
```

```python
fig, axs = plt.subplots(1, 2, layout='constrained')
axs[0].plot([1,2,3], [0,.5,.2])
axs[1].plot([3,2,1], [0,.5,.2])
fig.suptitle('Explicit interface')

for i in range(2):
    if i == 0:
        axs[i].set_xlabel('Woo')
    else:
        axs[i].set_xlabel('Boo')
```

```python
class DataContainer:
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def plot(self, ax=None, **kwargs):
        if ax is None:
            ax = plt.gca()
            
        ax.plot(self.x, self.y, **kwargs)
        ax.set_title('Plotted from DataClass')
        return ax
    
data = DataContainer([0, 1, 2, 3], [0, 0.2, 0.5, 0.3])
data.plot()
```
